```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
OJIP = pd.read_excel('Compiled_Sorghum_OJIP_Ace_2020.xlsx', sheet_name = 'Compiled_Data_2020_Raw')
```

```python
## Adds a leading zero if the plot number is not 4 digits
OJIP['Plot'] = OJIP['Plot'].astype(str).str.zfill(4)
```

```python
OJIP['Full_Date'] = pd.to_datetime(OJIP['Full_Date']).dt.date
```

```python
OJIP['Irrigation'] = OJIP
```

```python
sns.boxplot(data = OJIP, x = 'Date', y = 'Fv/Fm', hue = 'Irrigation');
```

```python
POI_WL = ['PI 677911', 'PI 678037', 'PI 678052', 'PI 678040']
POI_WW = ['PI 678305', 'PI 678248']
```

```python
g = sns.FacetGrid(data = OJIP, col = 'Full_Date', col_wrap=4, hue = 'Irrigation', aspect = 2, legend_out = True)
g.map(sns.scatterplot, 'Accession', 'Fv/Fm')
g.set_xticklabels(rotation = 45)
g.fig.tight_layout()
plt.legend();
```
